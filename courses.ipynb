```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
CoursesH = requests.get('https://cs.nyu.edu/dynamic/courses/schedule/?semester=spring_2021')
Coursedom = BeautifulSoup(CoursesH.text)
selector = '.schedule-listing .row'
CourseEle = Coursedom.select(selector)
df = pd.DataFrame(columns=["Number-Section", "Name","Instructor","Time"])
for element in CourseEle:
    spanfind = element.select('span')
    NumberSection = spanfind[0].text.split('(')[0].strip().split('\n')[0]
    Instructor = spanfind[2].text.strip().split('\n')[0]
    Instructor = Instructor.strip();
    Name = spanfind[1].text.strip();
    Time = spanfind[3].text.strip();
    df.loc[len(df.index)] = [NumberSection,Name,Instructor,Time]
df['Number'] = df['Number-Section'].str.extract('([^ ]+)..\d\d\d$')
df['Section'] = df['Number-Section'].str.extract('[^\.\d](\d\d\d)')
df.info()
```

```python

CoursesCata = requests.get('https://cs.nyu.edu/dynamic/courses/catalog/')
Catadom = BeautifulSoup(CoursesCata.text)
selector = '.col-sm-12'
CataEle = Catadom.select(selector)
dfCata = pd.DataFrame(columns=["Number", "Prereqs","Points"])
for element in CataEle:
    pfind = element.select('p')
    NumberC = pfind[0].text.strip().split(' ')[0]
    Points = pfind[1].text.split(".")[0].split("\n")[1].strip()
    Prereqs = pfind[2].text.split("Prerequisites: ")[1]
    dfCata.loc[len(dfCata.index)] = [NumberC,Prereqs,Points]
dfCata.info()
```

```python
df1 = df.drop(columns=['Number-Section','Section'])
dfnew = df1.merge(dfCata, how = 'left')
dfnew = dfnew[['Number', 'Name', 'Instructor', 'Time', 'Prereqs','Points']]
dfnew
```

I spot that sometimes there exists mutiple instructors responsible for one course. The way I coped with it is that I only count the first instructor.

I spot that sometimes one courses will have several different numbers(i.e. CSCI-GA.1170-002 DS-GA.1170). I only count the first name, which is the CSCI number. 

When I use pd.merge and how = left, which will keep all rows in the first DataFrame(df1, which is dataframe for the first website Spring 2021 CS Course schedule). This is important since we need to show all scheduled classes in spring 2021 along with their points and prereqs. The dfCata only represents the points and prereqs, while df1 reflects all scheduled classes. So we need how = left to keep  all rows in df1.
